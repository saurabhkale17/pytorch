```python
import torch
```

```python
x = torch.rand((2, 3))
print(x)
# Permutationof tensors
print(torch.einsum('ij->ji', x))
```

```python
# summantion
print(torch.einsum('ab->', x))
```

```python
# col sum
print(torch.einsum('ij->j', x))
# row sum
print(torch.einsum('ij->i', x))
```

```python
# matrix vector multiplication
v = torch.rand((1, 3))
x = torch.randn((2, 3))
# i=2, j=3, k=1
print(torch.einsum('ij,kj->ik', x, v))
```

```python
# matrix matrix multiplication
x = torch.randn((2, 3))
print(torch.einsum('ij,kj->ik', x, x))
print(torch.einsum('ij,ik->jk', x, x))
```

```python
# dot product of first row with forst row of matrix
torch.einsum('i, i->', x[0], x[0])
```

```python
# hadamard product element-wise multiplication
print(x)
torch.einsum('ij, ij->ij', x, x)
```

```python
a = torch.rand((3))
b = torch.randn((5))
print("a : ",a)
print("b : ", b)
print(torch.einsum('i, j -> ij', a, b))
```

```python
# batch  matrix multiplication
a = torch.rand((3, 2, 5))
b = torch.rand((3, 5, 3))
print(torch.einsum('ijk, ikl->ijl', a, b))
```
